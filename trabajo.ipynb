```python
import pandas as pd
!pip install mapclassify
import mapclassify
from sklearn.preprocessing import RobustScaler
df_educacion = pd.read_csv('C:/Users/USUARIO/Pictures/Saved Pictures/Education.csv')
df_electricidad = pd.read_csv('C:/Users/USUARIO/Pictures/Saved Pictures/Electricity.csv')
df_energia = pd.read_csv('C:/Users/USUARIO/Pictures/Saved Pictures/Energy.csv')
df_reservas = pd.read_csv('C:/Users/USUARIO/Pictures/Saved Pictures/Reserves.csv')
```

```python
# Nombres de columnas para inspeccion
print("Columnas df_educacion:", df_educacion.columns)
print("Columnas df_reservas:", df_reservas.columns)
print("Columnas df_electricidad:", df_electricidad.columns)
print("Columnas df_energia:", df_energia.columns)
```

```python
# Limpieza y formateo de electricidad reservas
df_reservas['value'] = df_reservas['value'].astype(str).str.strip()
df_reservas['value'] = df_reservas['value'].str.replace('$', '',)
df_reservas['value'] = df_reservas['value'].str.replace(',', '')
df_reservas['value'] = df_reservas['value'].str.strip()
df_reservas['reserves'] = pd.to_numeric(df_reservas['value'], errors='coerce')
```

```python
# Limpieza y formateo de electricidad
df_electricidad[' kW'] = df_electricidad[' kW'].str.replace(',', '')  # Reemplazar comas por puntos (decimales)
df_electricidad['electricity'] = pd.to_numeric(df_electricidad[' kW'], errors='coerce')
df_electricidad = df_electricidad.drop(columns=[' kW'])
```

```python
df_electricidad
```

```python
# Limpieza y formateo de electricidad de energÃ­a
df_energia[' Btu/person'] = df_energia[' Btu/person'].str.replace(',', '')
df_energia['energy'] = pd.to_numeric(df_energia[' Btu/person'], errors='coerce')
df_energia = df_energia.drop(columns=[' Btu/person'])
```

```python
df_energia
```

```python
#Renombramos la columna de educacion
df_educacion.rename(columns={'% of GDP': 'education'}, inplace=True)
```

```python
df_educacion
```

```python
#Seleccion de columnas qwe quiero
df_reservas = df_reservas[['name', 'region', 'reserves']]
df_educacion = df_educacion[['name', 'region', 'education']]
df_energia = df_energia[['name', 'region', 'energy']]
df_electricidad = df_electricidad[['name', 'region', 'electricity']]
```

```python
df_educacion
```

```python
#Combinamos los datos
df_combinado = pd.merge(df_reservas, df_energia, on=['name', 'region'], how='inner')
df_combinado = pd.merge(df_combinado, df_electricidad, on=['name', 'region'], how='inner')
df_combinado = pd.merge(df_combinado, df_educacion, on=['name', 'region'], how='inner')
```

```python
df_combinado
```

```python
#sacamos la media, min y maximo de datos por region
resumen_region = df_combinado.groupby('region')[['reserves', 'energy', 'electricity', 'education']].agg(['mean', 'min', 'max'])
resumen_region
```

```python
# Agrupamos por intervalos
#Rescalado de datos gigantes (energia,electricida) y usamos Quantiles
Vars = ['reserves', 'energy', 'electricity']
scaler = RobustScaler(unit_variance=True, with_centering=False)
robScaled_data = scaler.fit_transform(df_combinado[vars])
vars_rob_names = [v + '_rob' for v in vars]
df_combinado[vars_rob_names] = robScaled_data
for var_rob in vars_rob_names:
    q5 = mapclassify.Quantiles(df_combinado[var_rob], k=5)
    col_name = var_rob.replace('_rob', '') + '_level'
    df_combinado[col_name] = q5.yb + 1 
```

```python
#educacion clasificacion
K = 5
q5 = mapclassify.Quantiles(df_combinado['education'], k=K)
df_combinado['education_level'] = q5.yb + 1 
```

```python
#Definimos una varibles con las columnas obtenidas
df_final = df_combinado[[
    'name', 'region',
    'reserves', 'energy', 'electricity', 'education',
    'reserves_level', 'energy_level', 'electricity_level', 'education_level'
]]
```
